```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cars = pd.read_csv("CarPrice_Assignment.csv")
cars = cars.drop(['car_ID','CarName'],axis=1)

for c in cars.columns:
    if cars[c].dtype=='object':
        cars[c]=pd.factorize(cars[c])[0]

X=cars.drop('price',axis=1).values
y=cars['price'].values.reshape(-1,1)

X=(X-np.mean(X,axis=0))/np.std(X,axis=0)
X=np.hstack((np.ones((X.shape[0],1)),X))

np.random.seed(1)
idx=np.random.permutation(len(X))
train=int(0.8*len(X))
X_train,X_test=X[idx[:train]],X[idx[train:]]
y_train,y_test=y[idx[:train]],y[idx[train:]]

def grad_desc(X,y,lr=0.01,epochs=1000):
    m=len(y)
    th=np.zeros((X.shape[1],1))
    loss=[]
    for i in range(epochs):
        pred=X.dot(th)
        err=pred-y
        grad=(1/m)*X.T.dot(err)
        th-=lr*grad
        loss.append(np.mean(err**2))
    return th,loss

th,loss=grad_desc(X_train,y_train,0.01,1000)
pred=X_test.dot(th)

mse=np.mean((pred-y_test)**2)
r2=1-(np.sum((y_test-pred)**2)/np.sum((y_test-np.mean(y_test))**2))
print("MSE:",mse)
print("R2:",r2)

plt.plot(loss)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss curve")
plt.show()

plt.scatter(y_test,pred,alpha=0.6)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```
